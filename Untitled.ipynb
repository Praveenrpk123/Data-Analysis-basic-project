# Calculating the bore and the stroke

#Normally bore and stroke depends on the engine size and no of the cylinders in the engive whether it may two,four,six,five etc.

```python
df[df['bore'].isnull()] #Here value of the bore and the stroke are missing
```

#The formula for finding the value of bore and stroke is derived from the engine-size
Engine size=((bore*stroke)*2.54)*no.of cylinders.
By using the box plot we can find which should be calculated first

```python
plt.figure(figsize=(10,10),dpi=50)
print(sns.boxplot(df['bore']))
plt.figure(figsize=(10,10),dpi=50)
sns.boxplot(df['stroke'])
```

#with the help of the above box plots we can determine that stroke has more outliers compared to the bore
so we will calculate the bore first for getting an aaccurate value

```python
sns.distplot(df['bore'])
df['bore'].median()    #using the distplot
```

```python
df['bore']=df['bore'].fillna(df['bore'].mean())
df['bore']
df[df['stroke'].isnull()]   #filling the value of 'NaN' with the mean where median is the exact centre value where we use it for great accuracy
```

```python
sns.distplot(df['bore'])
```

```python
df['stroke']=df['stroke'].fillna(df['stroke'].median())
```

```python
df['stroke'].isnull().sum() #Hence we calculated it
```

# Analysis

```python
num_col=[]     #For analysing the numeric values
for i in df.columns:
    if (df[i].dtypes == 'int64') or (df[i].dtypes == 'float64'):
        num_col += [i]
categ_col=[]   #For analysing the Categorical values
for i in df.columns:
    if(df[i].dtypes == 'object'):
        categ_col += [i] 
```

#Analysing with the help of the distplot

```python
fig = plt.figure(figsize=(18,18))
for i,j in zip(range(1,len(num_col)+1),num_col):
    ax = fig.add_subplot(4,4,i)
    sns.distplot(df[j],ax=ax)
plt.show()
plt.savefig('dist.png')  #saving the figure as dist.png
```

```python
fig = plt.figure(figsize=(18,18))
for i,j in zip(range(1,len(categ_col)+1),categ_col):
    ax = fig.add_subplot(4,4,i)
    sns.countplot(df[j],ax=ax)
plt.show()
plt.savefig('count.png')  #saving the count plot figue as count.png
```

```python
fig = plt.figure(figsize=(18,18))
for i,j in zip(range(1,len(num_col)+1),num_col):
    ax = fig.add_subplot(4,4,i)
    sns.boxplot(y=df[j],ax=ax) #with the help of boxplot we can understand more about data related to outliers
plt.show()
```
